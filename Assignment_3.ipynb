```python
## Exercise 1: Prime Number Check
```

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

num = 29
print(f"{num} is prime:", is_prime(num))
```

```python
## Exercise 2: Product of Random Numbers
```

```python
import random

a = random.randint(1, 10)
b = random.randint(1, 10)

print("Numbers:", a, b)

user_answer = a * b
print("User Answer:", user_answer)

if user_answer == a * b:
    print("Correct Answer!")
else:
    print("Wrong Answer!")
```

```python
## Exercise 4: Word Counter
```

```python
text = "This is a sample text. This text will be used to demonstrate the word counter."
words = text.split()

word_count = {}
for word in words:
    word_count[word] = word_count.get(word, 0) + 1

print(word_count)
```

```python
## Exercise 5: Palindrome Check|
```

```python
def is_palindrome(s):
    s = ''.join(char.lower() for char in s if char.isalnum())
    return s == s[::-1]

test_str = "racecar"
print(f"{test_str} is palindrome:", is_palindrome(test_str))
```
